```python
resources = {
    "water": 800,
    "milk": 800,
    "coffee": 400,
}

Menu = {
    "espresso": {
        "ingredients": {"water": 50, "coffee": 18},
        "cost": 10,
    },
    "latte": {
        "ingredients": {"water": 200, "milk": 150, "coffee": 24},
        "cost": 15,
    },
    "cappuccino": {
        "ingredients": {"water": 250, "milk": 100, "coffee": 24},
        "cost": 20,
    }
}

money_collected = 30

while True:
    print("\nWhat would you like?")
    print("1. Espresso")
    print("2. Latte")
    print("3. Cappuccino")
    print("4. Report")
    print("5. Off")

    choice = input("Enter your choice: ")

    if choice == "5":
        print("Turning off the machine. Goodbye!")
        break

    elif choice == "4":
        print("\n----- Report -----")
        print(f"Water: {resources['water']}ml")
        print(f"Milk: {resources['milk']}ml")
        print(f"Coffee: {resources['coffee']}g")
        print(f"Money: ₹{money_collected}")
        continue

    elif choice == "1":
        print("\nYou selected: Espresso")
        drink = Menu["espresso"]
        ingredients = drink["ingredients"]
        cost = drink["cost"]

        if resources["water"] < ingredients["water"]:
            print("Sorry, not enough water.")
            continue
        if resources["coffee"] < ingredients["coffee"]:
            print("Sorry, not enough coffee.")
            continue

        amount = int(input("Please insert money (₹): "))

        if amount < cost:
            print("Sorry, that's not enough money. Money refunded.")
            continue
        elif amount > cost:
            balance = amount - cost
            print(f"Here is your espresso . Enjoy!")
            print(f"Here is ₹{balance} in change.")
        else:
            print("Here is your espresso . Enjoy!")

        resources["water"] -= ingredients["water"]
        resources["coffee"] -= ingredients["coffee"]
        money_collected += cost

    elif choice == "2":
        print("\nYou selected: Latte")
        drink = Menu["latte"]
        ingredients = drink["ingredients"]
        cost = drink["cost"]

        if resources["water"] < ingredients["water"]:
            print("Sorry, not enough water.")
            continue
        if resources["milk"] < ingredients["milk"]:
            print("Sorry, not enough milk.")
            continue
        if resources["coffee"] < ingredients["coffee"]:
            print("Sorry, not enough coffee.")
            continue

        amount = int(input("Please insert money (₹): "))

        if amount < cost:
            print("Sorry, that's not enough money. Money refunded.")
            continue
        elif amount > cost:
            balance = amount - cost
            print(f"Here is your latte . Enjoy!")
            print(f"Here is ₹{balance} in change.")
        else:
            print("Here is your latte . Enjoy!")

        resources["water"] -= ingredients["water"]
        resources["milk"] -= ingredients["milk"]
        resources["coffee"] -= ingredients["coffee"]
        money_collected += cost

    elif choice == "3":
        print("\nYou selected: Cappuccino")
        drink = Menu["cappuccino"]
        ingredients = drink["ingredients"]
        cost = drink["cost"]

        if resources["water"] < ingredients["water"]:
            print("Sorry, not enough water.")
            continue
        if resources["milk"] < ingredients["milk"]:
            print("Sorry, not enough milk.")
            continue
        if resources["coffee"] < ingredients["coffee"]:
            print("Sorry, not enough coffee.")
            continue

        amount = int(input("Please insert money (₹): "))

        if amount < cost:
            print("Sorry, that's not enough money. Money refunded.")
            continue
        elif amount > cost:
            balance = amount - cost
            print(f"Here is your cappuccino . Enjoy!")
            print(f"Here is ₹{balance} in change.")
        else:
            print("Here is your cappuccino . Enjoy!")

        resources["water"] -= ingredients["water"]
        resources["milk"] -= ingredients["milk"]
        resources["coffee"] -= ingredients["coffee"]
        money_collected += cost

    else:
        print("Invalid input. Please choose a number between 1 and 5.")
```
